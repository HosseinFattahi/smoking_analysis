```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'world_smoking_history_1924_2023.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
descriptive_stats = data.describe()
print(descriptive_stats)
```

```python
sns.set(style="whitegrid")

# Smoking Population Percentage over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Smoking_Population_Percentage', data=data)
plt.title('Smoking Population Percentage Over the Years')
plt.xlabel('Year')
plt.ylabel('Smoking Population Percentage')
plt.show()

# Male vs. Female Smokers Percentage
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Male_Smokers_Percentage', data=data, label='Male Smokers Percentage')
sns.lineplot(x='Year', y='Female_Smokers_Percentage', data=data, label='Female Smokers Percentage')
plt.title('Male vs. Female Smokers Percentage Over the Years')
plt.xlabel('Year')
plt.ylabel('Smokers Percentage')
plt.legend()
plt.show()

# Smoking Related Deaths over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Smoking_Related_Deaths', data=data)
plt.title('Smoking Related Deaths Over the Years')
plt.xlabel('Year')
plt.ylabel('Smoking Related Deaths')
plt.show()

# Healthcare Costs over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Healthcare_Costs', data=data)
plt.title('Healthcare Costs Over the Years')
plt.xlabel('Year')
plt.ylabel('Healthcare Costs')
plt.show()

# Anti-Smoking Campaigns over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Anti_Smoking_Campaigns', data=data)
plt.title('Anti-Smoking Campaigns Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Anti-Smoking Campaigns')
plt.show()
```

```python
numeric_columns = ['Smoking_Population_Percentage', 'Male_Smokers_Percentage', 'Female_Smokers_Percentage',
                   'Smoking_Related_Deaths', 'Age_Limit', 'Death_History', 'Average_Cigarettes_Per_Day',
                   'Healthcare_Costs', 'Anti_Smoking_Campaigns', 'Legislation_Strength', 'Youth_Smokers_Percentage']

# Calculate correlation matrix
correlation_matrix = data[numeric_columns].corr()

# Plot correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
